In this notebook we give an application of our theorem on the marginal distribution of individual conditional distributions with polynomial tails, where the probability of sampling from a given individual distribution is uniform

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def pickands_estimator(x,thresh,avg_size):
    y=x.copy()
    length=x.shape[0]
    cutoff=int(thresh*length)
    x=-np.sort(-np.sort(x, axis=0)[cutoff:], axis=0)
    ksi=np.array([])
    for j in np.arange(np.sqrt(length-cutoff))[-avg_size:]:
        k=int(j+1)
        ksi=np.append(ksi,np.log((x[k]-x[2*k])/(x[2*k]-x[4*k]))/np.log(2))
    ksi=ksi.reshape(avg_size,-1)
    ksi=ksi.mean(0)
    return ksi
```

```python
avg_hist=[]
std_hist=[]
n=5000000
#For each max_ksi_j in the array 'np.linspace(-5,5,50)', we will perform the following process:
#we will sample 'n' samples from each of 20 distributions. These distributions are constructed such that 
#their tails have shape parameters as in the following array 'np.linspace(-5,max_ksi_j,20)' respectively.
#The sample marginal (mixture) distribution (denote by 'M') of these 20 distributions, consists of the array that
#contain all '20*n' samples. We calculate the shape parameter of the tail of this marginal distribution 'M'.
#This should be very close to max_ksi_j
#We run the experiment 10 times to study the variability of predictions
for max_ksi in np.linspace(-5,5,50):
    print(max_ksi)
    run_hist=[]
    for run in range(10):
        marginal_sample=np.array([])
        print('run:', run)
        for ksi in np.linspace(-5,max_ksi,20):

            x=np.random.rand(n)
            sample=x**(-ksi)
            marginal_sample=np.append(marginal_sample,sample)

        run_hist.append(pickands_estimator(marginal_sample,0.995,5))
    avg_hist.append(np.array(run_hist).mean())
    std_hist.append(np.array(run_hist).std())
```

```python
avg_hist=np.array(avg_hist)
std_hist=np.array(std_hist)
```

```python
plt.plot(np.linspace(-5,5,50),np.linspace(-5,5,50))
plt.plot(np.linspace(-5,5,50),avg_hist)
plt.plot(np.linspace(-5,5,50),avg_hist-3*std_hist)
plt.plot(np.linspace(-5,5,50),avg_hist+3*std_hist)
```

```python
xi_list=np.linspace(-5,5,50)
```

```python
def trunc(values, decs=0):
    return np.trunc(values*10**decs)/(10**decs)
```

```python
predictions=avg_hist
deviations=std_hist
```

```python
#Illustrate results
figure(figsize=(10, 10), dpi=120)
plt.plot(xi_list[xi_list>0],xi_list[xi_list>0],color='red', label='Theoretical Prediction')
plt.fill_between(xi_list[xi_list<=0],predictions[xi_list<=0]*0,predictions[xi_list<=0]*0-5,color='red', alpha=0.2,label='Theoretical Prediction')
plt.plot(xi_list,predictions,color='blue', label='The average of shape parameter estimations')
plt.plot(xi_list,predictions-3*deviations,color='orange', label='Three standard deviations')
plt.plot(xi_list,predictions+3*deviations,color='orange')
plt.fill_between(xi_list,predictions-3*deviations,predictions+3*deviations)
plt.xlabel('Maximal Shape Parameter of Individual Distributions')
xticks_as_strings=trunc(xi_list, decs=2).astype(str)
plt.xticks(ticks=xi_list, labels=xticks_as_strings)
plt.xticks(rotation=90)
plt.ylabel('Average of Estimated Shape Parameter Values of the Marginal Distribution')
yticks_as_strings=trunc(xi_list, decs=2).astype(str)
plt.yticks(ticks=xi_list, labels=yticks_as_strings)
plt.yticks(rotation=0)
plt.legend()
plt.savefig('theorem_experiment1_uniform.jpg')
```
